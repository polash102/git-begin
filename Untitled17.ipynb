```python
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory("/content/drive/MyDrive/MangoLeafBD Dataset")

ds_train=tf.keras.preprocessing.image_dataset_from_directory(
    '/content/drive/MyDrive/MangoLeafBD Dataset',validation_split=0.2,subset='training',seed=123)

ds_test=tf.keras.preprocessing.image_dataset_from_directory(
    '/content/drive/MyDrive/MangoLeafBD Dataset',validation_split=0.2,subset='validation',seed=123)
```

```python
# Function to preprocess dataset
def prepare_data(ds):
    images, labels = [], []
    for img, lbl in tfds.as_numpy(ds):
        # Resize, normalize, and flatten images
        img = tf.image.resize(img, (128, 128)).numpy().astype(np.float32) / 255.0  # Normalize
        images.append(img.flatten())  # Flatten image
        labels.append(lbl)

    return np.stack(images), np.array(labels)  # Ensure uniform shape
```

```python
# Load dataset (Example: MNIST or CIFAR-10)
ds_train, ds_test = tfds.load('mnist', split=['train', 'test'], as_supervised=True)
```

```python
# Prepare training and testing data
X_train, y_train = prepare_data(ds_train)
X_test, y_test = prepare_data(ds_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
# Get class names from the original dataset loaded in the first cell
class_names = dataset.class_names  # Use 'dataset' instead of 'ds_train'

for images,labels in dataset.take(1): # Use 'dataset' instead of 'ds_train'
  for i in range(len(class_names)): # Loop through the number of classes you have
    ax=plt.subplot(3,3,i+1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train Decision Tree Model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, y_pred_dt)
```

```python
# Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)
```

```python
# Compare Models
print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")

if rf_accuracy > dt_accuracy:
    print("✅ Random Forest performs better!")
else:
    print("✅ Decision Tree performs better!")
```
